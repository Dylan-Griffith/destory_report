```python
import pandas as pd
import datetime

"""
Will search through recrods spreadsheets and find boxes
that need to be pulled for destruction
"""

today = datetime.datetime.now()

def find_boxes(filename):
    """
    creates dataframe of boxes to destroy
    returns dataframe object
    """
    df = pd.read_excel(filename)
    
    # Filtering based on not destroyed
    not_dest = df['Destroyed?'] == 'No'
    df = df.loc[not_dest]
    
    # Filtering based on Perm
    no_perm = df['Destroy Date'] != 'Perm'
    df = df.loc[no_perm]
    
    # removing all empty destroy date rows
    df.dropna(subset=['Destroy Date'], inplace=True)
    df.dropna(subset=['Box Loc.'], inplace=True)
    
    to_be_destroyed = df.loc[(df["Destroyed?"] == 'No') & (df['Destroy Date'] < today)]
    
    return to_be_destroyed

def main():
    rec2 = find_boxes('recctrtwo_.xlsx')
    rec3 = find_boxes('recctrthree (Recovered) (Autosaved)_.xlsx')
    
    final_df = pd.concat([rec2,rec3])
    final_df.set_index('Box No.', inplace=True)
    final_df.sort_values('Department', inplace=True)
    
    final_df.to_csv('to_be_destroyed.csv')
    
    print('Program Complete')

if __name__ == '__main__':
    main()
```
